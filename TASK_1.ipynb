Import Libraries & Load the Dataset

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("netflix_sample_dirty.xlsx")

# Display first few rows
df.head()
```

Identify and Handle Missing Values

```python
# Show count of missing values per column
print(df.isnull().sum())

# Optionally fill or drop missing values
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Not Available", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['gender'].fillna("Unknown", inplace=True)
df.fillna("Unknown", inplace=True)  # Fills remaining NaNs
```

Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

Standardize Text Values

```python
# Gender normalization
df['gender'] = df['gender'].str.lower().replace({
    'm': 'male', 'f': 'female', 'unknown': 'unknown'
})

# Country normalization
df['country'] = df['country'].replace({
    "U.S.A": "USA", "United States": "USA", "Uk": "UK"
})
```

Convert Date Formats

```python
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Optional: Convert to specific format
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

Clean Column Names

```python
# Rename columns: lowercase & replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

Fix Data Types

```python
# Convert release_year to numeric
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')

# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```
